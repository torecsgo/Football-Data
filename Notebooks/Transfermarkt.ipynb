```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException

from datetime import date
import pandas as pd
import time

c_options = webdriver.ChromeOptions()
c_options.add_argument("--start-maximized")
c_driver = webdriver.Chrome(options=c_options)

c_driver.get('https://www.transfermarkt.com/')
```

```python
url = "https://www.transfermarkt.com/jumplist/marktwerte/wettbewerb/"
top_leagues = {
    'Premier League': 'GB1',
    'LaLiga': 'ES1',
    'Bundesliga': 'L1',
    'Serie A': 'IT1',
    'Ligue 1': 'FR1'
}

# try:
#     cookies_button = WebDriverWait(c_driver, 15).until(
#         EC.element_to_be_clickable((By.XPATH, '/html/body/div/div[2]/div[3]/div[1]/div/button'))
#     )
#     cookies_button.click()
#     print("Cookies aceptadas.")
# except TimeoutException:
#     print("El botón de cookies no apareció en el tiempo esperado.")
```

```python
news_list = []
for k, v in top_leagues.items():
    c_driver.get(url + v)
    time.sleep(2)

    #/html/body/div/main/div[1]/div[1]/div/div[3]/div/table/tbody/tr[1]
    for page in range(1, 5):
        players_table = '/html/body/div/main/div[1]/div[1]/div/div[3]/div/table/tbody/tr'
        players = c_driver.find_elements(By.XPATH, players_table)
        for player in players:
            player_info = {
                'date': date.today(),
                'league': k,
                'ranking': player.find_element(By.XPATH, './td[1]').text,
                'name': player.find_element(By.XPATH, './td[2]/table/tbody/tr[1]/td[2]/a').text,
                'position': player.find_element(By.XPATH, './td[2]/table/tbody/tr[2]/td').text,
                'nationality': player.find_element(By.XPATH, './td[3]/img[1]').get_attribute('title'),
                'age': player.find_element(By.XPATH, './td[4]').text,
                'club': player.find_element(By.XPATH, './td[5]/a/img').get_attribute('title'),
                'value': player.find_element(By.XPATH, './td[6]/a').text
            }
            news_list.append(player_info)
        new_page = c_driver.find_element(By.XPATH, f'/html/body/div/main/div[1]/div[1]/div/div[3]/div/div[2]/ul/li[{page + 1}]/a').get_attribute('href')
        c_driver.get(new_page)
        time.sleep(5)

print("FIN")
```
